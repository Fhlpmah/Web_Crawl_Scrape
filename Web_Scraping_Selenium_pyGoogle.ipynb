```python
from bs4 import BeautifulSoup as bs
import requests
import time
import pandas as pd
import csv
```

Let's look at the format of the website first. <br>
<a href="http://quotes.toscrape.com/"> Quotes! </a>

```python
text_list = []
author_list = []

for i in range(1, 15):
    url = f"https://quotes.toscrape.com/page/{i}/"
    link = requests.get(url)
    html = link.text
    soup = bs(html, 'html.parser')

    texts = soup.find_all('span', itemprop='text')
    authors = soup.find_all('small', itemprop = 'author')

    for text in texts:
        text_list.append(text.get_text())

    for author in authors:
        author_list.append(author.get_text())

    df1 = pd.DataFrame(data=text_list, columns=['Text of Quote'])
    df2 = pd.DataFrame(data=author_list, columns=['Author'])
```

Let's check each df

```python
df1
```

```python
df2
```

Let's combine the dataframes together and export it

```python
texts = pd.concat([df1, df2], axis=1)
```

```python
def get_title(search):
    """"fetching the titles of each article"""
    stories = []
    search_key = gn.search(search)
    result = gn.result(search_key)
    for item in result:
        story = {'title': item['title'],
                'link': item['link']
                }
        stories.append(story)
    return stories

a = get_title('weather')
```
